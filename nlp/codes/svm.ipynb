```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.pipeline import make_pipeline


file_path = ''
data = pd.read_excel(file_path)


data.columns = data.columns.str.strip()


data = data.dropna(subset=['Sentence', 'Label'])
data['Sentence'] = data['Sentence'].astype(str)
data['Label'] = data['Label'].astype(str)

data['Label'] = data['Label'].apply(lambda x: 1 if x.strip() == 'Formal' else 0)


X_train, X_test, y_train, y_test = train_test_split(
    data['Sentence'], data['Label'], test_size=0.3, random_state=42
)


model = make_pipeline(TfidfVectorizer(), SVC())

# Training the model
model.fit(X_train, y_train)

# Evaluating the model
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.pipeline import make_pipeline

file_path = '/content/combined nlp.xlsx'
data = pd.read_excel(file_path)


data.columns = data.columns.str.strip()


data = data.dropna(subset=['Sentence', 'Label'])
data['Sentence'] = data['Sentence'].astype(str)
data['Label'] = data['Label'].astype(str)


data['Label'] = data['Label'].apply(lambda x: 1 if x.strip() == 'Formal' else 0)

X_train, X_test, y_train, y_test = train_test_split(
    data['Sentence'], data['Label'], test_size=0.1, random_state=42
)


model = make_pipeline(TfidfVectorizer(), SVC())


model.fit(X_train, y_train)


predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```

```python

def predict_sentence(sentence, model):

    sentence = str(sentence)

    prediction = model.predict([sentence])[0]

    return "Formal" if prediction == 1 else "Informal"

new_sentence = "రకాలైన services ను అందించగలదు, కంపోజ్"
prediction = predict_sentence(new_sentence, model)
print(f"The sentence is predicted to be: {prediction}")
```

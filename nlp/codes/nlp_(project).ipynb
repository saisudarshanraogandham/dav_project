```python
import openai

def convert_to_informal(formal_sentence):
    try:
        response = openai.Completion.create(
            engine="text-davinci-003",
            prompt=f"Rewrite the following sentence in a more informal and casual way:\n\n'{formal_sentence}'\n\nInformal version:",
            max_tokens=60
        )
        informal_sentence = response.choices[0].text.strip()
        return informal_sentence
    except Exception as e:
        print(f"An error occurred: {e}")
        return None


formal_sentence = "It is imperative that we find a solution to this problem immediately."
informal_sentence = convert_to_informal(formal_sentence)
print(informal_sentence)
```
